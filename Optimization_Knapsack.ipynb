### Data Preprocessing and Imports

```python
# Import statements
import mlrose
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
# Initialize fitness function object using pre-defined class
weights = [10, 5, 2, 8, 15]
values = [1, 2, 3, 4, 5]
max_weight_pct = 0.6
fitness = mlrose.Knapsack(weights, values, max_weight_pct)

# Define optimization problem object
problem = mlrose.DiscreteOpt(length = 5, fitness_fn = fitness, maximize=True, max_val=5)
```

```python
# Setup problem for training

max_iter = 1002
num_iter = range(int(1),int(max_iter), 20)
init_state = np.array([1, 0, 2, 1, 0])

```

### Randomized Hill Climbing

```python
# Solve using randomized hill climbing - attempt 1
np.random.seed(1)
rhc_performance = [0] * len(num_iter)
               
for i, o in enumerate(num_iter):
    print('maximizing fitness function with number of iterations' + str(o))               
    [best_state, best_fitness] = mlrose.random_hill_climb(problem, max_attempts = 10, 
                                                      max_iters = num_iter[i], restarts = 0, init_state = init_state)
    rhc_performance[i] = best_fitness
    print('fitness: ' + str(rhc_performance[i]))
    print('---')
```

### Simulated Annealing

```python
# Solve using simulated annealing
np.random.seed(1)

# Define decay schedule
schedule = mlrose.ExpDecay()
sa_performance = [0] * len(num_iter)
               
for i, o in enumerate(num_iter):
    print('maximizing fitness function with number of iterations' + str(o))               
    [best_state, best_fitness] = mlrose.simulated_annealing(problem, schedule = schedule, max_attempts = 10, 
                                                      max_iters = num_iter[i], init_state = init_state)
    sa_performance[i] = best_fitness
    print('fitness: ' + str(sa_performance[i]))
    print('---')
```

### Genetic Algorithms

```python
# Solve using genetic algorithms
np.random.seed(1)

ga_performance = [0] * len(num_iter)
               
for i, o in enumerate(num_iter):
    print('maximizing fitness function with number of iterations' + str(o))               
    best_state, best_fitness = mlrose.genetic_alg(problem, pop_size=200, mutation_prob=0.1, 
                                              max_attempts=10, max_iters=num_iter[i])
    ga_performance[i] = best_fitness
    print('fitness: ' + str(ga_performance[i]))
    print('---')
```

### MIMIC

```python
# Solve using MIMIC 
np.random.seed(1)

mimic_performance = [0] * len(num_iter)
               
for i, o in enumerate(num_iter):
    print('maximizing fitness function with number of iterations' + str(o))               
    best_state, best_fitness = mlrose.mimic(problem, pop_size=200, keep_pct=0.2, 
                                        max_attempts=10, max_iters=num_iter[i])
    mimic_performance[i] = best_fitness
    print('fitness: ' + str(mimic_performance[i]))
    print('---')
```

### Graph of Results

```python
    # Plot results
    print('plotting results')
    plt.figure()
    title = 'Knapsack: Performance x Iterations'
    plt.title(title)
    plt.plot(num_iter, rhc_performance, '-', label='RHC')
    plt.plot(num_iter, sa_performance, '-', label='SA')
    plt.plot(num_iter, ga_performance, '-', label='GA')
    plt.plot(num_iter, mimic_performance, '-', label='MIMIC')
    plt.legend()
    plt.xlabel('Number of Iterations')
    plt.ylabel('Best Fitness')
    print('plot complete')
    ### ---
```
